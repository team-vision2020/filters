```python
import invert
from skimage.io import imread, imsave
from skimage import img_as_float
from matplotlib import pyplot as plt
import matplotlib as mpl
```

```python
in_fn = "reyes.jpg"
```

```python
image = img_as_float(imread(in_fn))
```

```python
plt.imshow(image)
plt.show()
```

```python
model = invert.load_model("reyes.h5")
```

```python
result = invert.unfilter_image(image, model)
```

```python
plt.imshow(result)
plt.show()
```

```python
imsave("inverted-reyes.jpg", result)
```
